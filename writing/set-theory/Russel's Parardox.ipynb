A class and set are objects which can contain elements. A set is a special kind of class; one that is restricted in quality by the axioms of set theory to avoid paradoxes such as the one below. In the langauge of First Order Logic, classes can be constructed as follows:

Let $P(x)$ be a statement about $x$ which can be expressed from the symbols:

$\in$     -    'element of'

$\vee$     -    'and'

$\wedge$    -   'or'

$\Rightarrow$  'implies'
 
$\neg$    -     'not'

$\forall$   -   'for all'

Along with parentheses, brackets, and variables.

A class is a collection of elements which satisfy $P(x)$.

In other words, a class is a collection of all the objects $x$, such that plugging them into the function $P(x)$ gives back a value "True." In the langauge of first order logic this can be written like this:

$ [x | P(x)] $

One paradoxical class which can be constructed via the unprohibited nature of class cosnstruction is the subject of Russel's Paradox.

Given a class $C$, we define its members to be classes which are not elements of themselves. The most confusing concept to wrap one's mind around here is the idea of a class containing itself, but having an intuitive understanding will not be as important as carefully applying the definitions and following the implications of doing so, as is done below.

$ C = [ x | \neg (x \in x) ] $

*C is the set of all x such that x is not contained in x.*

An interesting question to ask is whether or not $C$ contains itself. With the principle of the excluded middle in mind (an accepted rule in the context of set theory stating that *a logical conclusion is either true or false - and no other combination like TT, FF, or neither*) we investigate by examining either conclusion:

Keep in mind that 'being a member of,' and 'being contained in' are exactly the same.

$C$ is a member of itself.
Since $C$ is defined as the class of classes which do not contain themselves, it must not actually be a member of itself like we just assumed.

So this either shows that $C$ is not a member of itself (assuming the law of excluded middle), or if that is also proven false that we have a paradox; investigating:

$C$ is not a member of itself.
By the definition of $C$, it contains all classes which are not members of themselves, therefore if $C$ is not self containing as we just asserted then it must be in $C$, which implies that it contains itself, contradicting the assumption that it is not a member of itself.

Two opposite statements result in conflicting conclusions, something about the lack or restrictions of class construction leads to this paradox (you'll have to research Set Theory to learn the specifics).

Symbollicly:

$ C \in C \leftrightarrow \neg(C \in C)$

*If $C$ contains $C$ then $C$ does not contain $C$, and vice versa.*

So a strange line of reasoning led to an contradiction - it isn't necessarily obvious why this is a big deal when any example like this could simply be labelled 'paradoxical' and just be considered a dead end. There is one more concept to explore called the principle of explosion, which sheds light on how paradoxes allowed to arise from a logical framework invalidates that framework.


Consider we have some rules which allow a paradox to exist:

*All apples are red.*

*Some apples are yellow.*

Each statement above is true despite contradicting eachother, because they are axioms. Paradoxes like this can easily become resultant of poorly formed sets of axioms (the ground truths of a theory).


Now we say:

*All apples are red, or unicorns exist on Earth.*

*Some apples are yellow, or unicorns exist on Earth.*

Because the first propisition in either statement is true, both statements as a whole are true. But because we can use either fact about apples to disprove the other apple claim, we can force the propisitions on the other side of the "or" to be true, i.e. unicorns exist on Earth. 

Because the statement "All apples are red, or unicorns exist on Earth" is true, and we can disprove the propisition about apples using our contradictory fact, we force the other claim to be true to satisfy the truth value we already placed on the whole statement.

Since this process can be done to prove any statement from a single contradiction, a formal system with this bug can no longer make meaningful, explicit claims. 

This fundamental flaw in what is now known as 'niave set theory' motivated the formal set theories we have today which introduce the notion of sets as opposed to classes - sets in set theory are specific types of classes with restrictions in place to avoid such paradoxes. Though less straightforward, the sets laid out in these theories give rise to previously disperate mathematics - unifying and justifying thousands of years of work into a single framework.

There is one more concept to explore called the *principle of explosion*, which sheds light on how paradoxes allowed to arise from a set of axioms make the theory built from those axioms effectively meaningless.


Consider we have some axioms which allow a paradox to exist:

*All apples are red.*

*Some apples are yellow.*

Because they are axioms, each statement above is true despite contradicting eachother. Paradoxes like this can easily become resultant of poorly formed sets of axioms, though will usually appear later on via deductions. The direct contradiction of the axioms themselves is just to make this example very clear.

Now we say:

*All apples are red, or unicorns exist on Earth.*

*Some apples are yellow, or unicorns exist on Earth.*

We have said that either one of our axioms is true, or the claim that unicorns exist on Earth, is true. As a whole, the new statement is true, and from our axioms seems to say that unicorns must not exist on Earth.

But because we can use either claim about apples to disprove the other, i.e. "Some apples are yellow" is false, because we know that "All apples are red", we can also change which statement is true within our "or" statement.

All we have to do is disprove "Some apples are yellow" as we did above, and now we can deduce from "Some apples are yellow, or unicorns exist on Earth" that unicorns must exist on earth, according to our axioms. Importantly, *any* claim can be subsituted for our one about unicorns and be proven in this manner, essentially making the theory built from those axioms completely meaningless, and by definition, paradoxical.
